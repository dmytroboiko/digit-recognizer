```python
from keras import backend as K
from keras.models import Sequential
from keras.utils import np_utils
from keras.layers.core import Dense, Activation, Dropout
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
compressed = np.load('data/compressed_images.npy')
weights = np.load('data/decoder_weights.npy')
biases = np.load('data/decoder_biases.npy')
```

```python
model = Sequential()
model.add(Dense(784, input_dim=64, weights=[weights, biases]))
model.add(Activation('tanh'))
```

```python
images = model.predict(compressed)
```

```python
fig = plt.figure(figsize=(18, 4))
for i in range(20):
    fig.add_subplot(2, 10, i + 1)
    plt.imshow(images[i].reshape(28,28), cmap='gray')
```
